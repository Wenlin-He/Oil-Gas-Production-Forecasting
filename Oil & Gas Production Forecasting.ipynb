# Assumptions made

## Logisitc regression:
- **Linearity Assumption**: Logistic regression assumes a linear relationship between the independent variables (x_var) and the log-odds of the dependent variable (LiquidsProd_BBL,GasProd_MCF) . If the relationship is highly non-linear, logistic regression may not perform well.

- **Feature Independence**: Logistic regression assumes that the independent variables are independent of each other. 
    - No Multicollinearity: The code utilizes ordinal encoding for categorical features and scales the highly skewed numerical features, which suggests an assumption of no multicollinearity (or high correlation) between independent variables.
We have plotted a heat map (correlation matrix) below to elimite the variables that are highly correlated (corr >0.8). If there are strong correlations between the independent variables (multicollinearity), it can lead to unstable estimates of the coefficients.

- **Feature engineering** (hydrocarbon_to_water_ratio) assumes that the engineered features are relevant to the prediction task. Irrelevant or redundant features may introduce noise and reduce the model's predictive performance.


## SARIMA Model Specific Assumptions:
- **Stationarity of Time Series Data:** Even after applying differencing to achieve stationarity, there's an assumption that the transformed series sufficiently meets the stationarity requirement for SARIMA modeling. The presence of non-stationary elements not addressed by differencing could undermine the model's validity.
- **Correct Identification of Seasonal Patterns:** The selection of seasonal parameters (P, D, Q, and m) assumes that the seasonal patterns within the data have been accurately identified and that these patterns will persist over the forecast period. Misidentification of the seasonal cycle length (m) or the degree of seasonal differencing (D) could lead to inadequate model fitting and forecasting errors.
- **Adequacy of Model Parameters:** The process of selecting the best SARIMA model parameters based on AIC values assumes that the model with the lowest AIC adequately captures the underlying process generating the time series data. However, this criterion may not always result in the best predictive model, especially if other model diagnostic checks (like residual analysis) are not thoroughly conducted.
- **Forecast Horizon Reliability:** There's an implicit assumption regarding the reliability of forecasts over the chosen forecast horizon. The farther out the forecast, the more uncertain it becomes, especially in volatile industries like oil and gas, where external factors can significantly impact production trends.


# Limitations



## Logistic Regression Limitations:

- There can be more features to be considered as critical other than those picked for x_var.

- **Binary Categorization of Production Data:** Converting continuous production data into binary categories based on the mean may oversimplify the complex nature of production outputs, potentially losing valuable information about the data's distribution and variability.

- **Predictor Selection:** The choice of predictors for the logistic regression model is based on the variables available in the dataset. This selection process may overlook other significant factors not captured in the dataset that could influence production, such as geological characteristics not explicitly measured or external economic factors.

- **Assumption of Independence:** The logistic regression model assumes that each data point (e.g., each well or production instance) is independent. In the context of oil and gas production, the performance of one well could be influenced by the operations of nearby wells, leading to spatial dependencies that the model does not account for.

- **Potential for Multicollinearity:** Given the use of multiple predictors, there's a risk of multicollinearity. This multicollinearity can inflate the variance of the coefficient estimates and make it difficult to interpret the effects of individual predictors.

## SARIMA Model Limitations:

- **Parameter Selection Process:** The process of selecting SARIMA model parameters (p, d, q, P, D, Q, m) based on AIC may not always yield the best predictive model. The reliance on AIC as the sole criterion for model selection overlooks other important factors such as model simplicity, interpretability, and out-of-sample forecast performance.

- **Forecast Horizon Uncertainty:** The forecasts generated by the SARIMA model are subject to increased uncertainty as the forecast horizon extends, particularly in an industry as volatile and susceptible to external influences as oil and gas. This can limit the model's practical applicability for long-term planning purposes.

```python
# Determined variable after the above EDA and data analysis
x_var =[
    'WaterProd_BBL',
    'TotalProdMonths',
    #'TVD_FT',
    'MD_FT',
    'Hydrocarbon_to_Water_Ratio', # Calculated Field
    #'Proppant_LBS',
    #'TotalWaterPumped_GAL'
    'TotalFluidPumped_BBL',
    'WellPadDirection',
    #'Field',
    #'StateWellType'
    
]
```

```python
# Select data for logistic regression
df_lg = merged_data.copy()
```

```python
x_data = df_lg.loc[:,x_var]

y_data_oil = df_lg.loc[:,'LiquidsProd_BBL_new']
y_data_gas = df_lg.loc[:,'GasProd_MCF_new']
```

```python
# Visualize encoded variables with its corresponding code (FOR INPUT)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_data['WellPadDirection_encoded'] = le.fit_transform(x_data['WellPadDirection'])
x_data.groupby('WellPadDirection')['WellPadDirection_encoded'].mean()
```

```python
#Ordinal Encoding
from sklearn.preprocessing import OrdinalEncoder

s = (x_data.dtypes == 'object')
object_cols = list(s[s].index)

ordinal_encoder = OrdinalEncoder()
x_data[object_cols] = ordinal_encoder.fit_transform(x_data[object_cols])
```

```python
# OIL - Split the data to train and test for the numerical variables targeting at 'y'
x_train_oil, x_test_oil, y_train_oil, y_test_oil = train_test_split(
            x_data, 
            y_data_oil, 
            test_size    = 0.25,
            random_state = 219)
```

```python
# GAS - Split the data to train and test for the numerical variables targeting at 'y'
x_train_gas, x_test_gas, y_train_gas, y_test_gas = train_test_split(
            x_data, 
            y_data_gas, 
            test_size    = 0.25,
            random_state = 219)
```

```python
# Calculate the skewness for all features
feature_skewness_oil = x_train_oil[x_var].skew().sort_values(ascending=False)
print(feature_skewness_oil)
# Select features with skewness above 1/below -1
highly_skewed_features_oil = feature_skewness_oil[abs(feature_skewness_oil) > 1].index
print(f"highly skewed features: {highly_skewed_features_oil.values}")

# Apply transformations (yeojohnson for highly skewed features)
from scipy.stats import yeojohnson
from sklearn.preprocessing import StandardScaler

# Standardize the features before applying Yeo-Johnson to reduce the range of values
scaler = StandardScaler()
x_train_oil[highly_skewed_features_oil] = scaler.fit_transform(x_train_oil[highly_skewed_features_oil])
x_test_oil[highly_skewed_features_oil] = scaler.transform(x_test_oil[highly_skewed_features_oil])

# Apply Yeo-Johnson transformation
for feature in highly_skewed_features_oil:
    # Transform the training data and store the lambda value
    x_train_oil[feature], fitted_lambda = yeojohnson(x_train_oil[feature])
    # Apply the same lambda value to transform the test data
    x_test_oil[feature] = yeojohnson(x_test_oil[feature], lmbda=fitted_lambda)[0]
```

```python
# Calculate the skewness for all features
feature_skewness_gas = x_train_gas[x_var].skew().sort_values(ascending=False)
print(feature_skewness_gas)
# Select features with skewness above 1/below -1
highly_skewed_features_gas = feature_skewness_gas[abs(feature_skewness_gas) > 1].index
print(f"highly skewed features: {highly_skewed_features_gas.values}")

# Apply transformations (yeojohnson for highly skewed features)
from scipy.stats import yeojohnson
from sklearn.preprocessing import StandardScaler

# Standardize the features before applying Yeo-Johnson to reduce the range of values
scaler = StandardScaler()
x_train_gas[highly_skewed_features_gas] = scaler.fit_transform(x_train_gas[highly_skewed_features_gas])
x_test_gas[highly_skewed_features_gas] = scaler.transform(x_test_gas[highly_skewed_features_gas])

# Apply Yeo-Johnson transformation
for feature in highly_skewed_features_gas:
    # Transform the training data and store the lambda value
    x_train_gas[feature], fitted_lambda = yeojohnson(x_train_gas[feature])
    # Apply the same lambda value to transform the test data
    x_test_gas[feature] = yeojohnson(x_test_gas[feature], lmbda=fitted_lambda)[0]
```

```python
# Check the skewness again after the transformation
feature_skewness= x_train_oil[x_var].skew().sort_values(ascending=False)
print(feature_skewness)
```

### Input for OIL production

**Input table for WellPadDirection**



```python
# logit(P(Y=1)) = 0.0007 + (-0.0069069210)*WaterProd_BBL + (0.1993890205)*TotalProdMonths + (-0.0000410087)*MD_FT + (-0.0000494911)*Hydrocarbon_to_Water_Ratio + (-0.0000055366)*TotalFluidPumped_BBL + (0.0014357425)*WellPadDirection
def classify_oil_production(WaterProd_BBL, TotalProdMonths, MD_FT, Hydrocarbon_to_Water_Ratio, TotalFluidPumped_BBL,WellPadDirection):
    # Coefficients from the formula
    intercept = 0.0007
    coef_WaterProd_BBL = -0.0069069210
    coef_TotalProdMonths = 0.1993890205
    coef_MD_FT = -0.0000410087
    coef_Hydrocarbon_to_Water_Ratio = -0.0000494911
    coef_TotalFluidPumped_BBL = -0.0000055366
    coef_WellPadDirection = 0.0014357425
    
    # Calculate the log-odds using the formula
    log_odds = intercept + \
                coef_WaterProd_BBL * WaterProd_BBL + \
                coef_TotalProdMonths * TotalProdMonths + \
                coef_MD_FT * MD_FT + \
                coef_Hydrocarbon_to_Water_Ratio * Hydrocarbon_to_Water_Ratio + \
                coef_TotalFluidPumped_BBL * TotalFluidPumped_BBL + \
                coef_WellPadDirection * WellPadDirection
    
    # Convert log-odds to probability
    probability = 1 / (1 + pow(2.71828, -log_odds))
    
    # Classify the production based on the probability
    if probability > 0.5:
        return "High oil production well"
    else:
        return "Low oil production well"

# Inputs:
WaterProd_BBL = 40000
TotalProdMonths = 1
MD_FT = 66268
Hydrocarbon_to_Water_Ratio = 0.076553
TotalFluidPumped_BBL = 20000
WellPadDirection = 1 # Input 0 for E, 1 for N, 2 for NE, 3 for NW, 4 for S, 5 for W

classification = classify_oil_production(WaterProd_BBL, TotalProdMonths, MD_FT, Hydrocarbon_to_Water_Ratio, TotalFluidPumped_BBL, WellPadDirection)
print(classification)
```

```python
import itertools
import statsmodels.api as sm
import math

# Define the range of values for p, d, q, P, D, Q, and m
p_values = range(0, 3)  # Autoregressive order
d_values = [0]          # Differencing order
q_values = range(0, 3)  # Moving average order
P_values = range(0, 2)  # Seasonal autoregressive order
D_values = range(0, 1)  # Seasonal differencing order
Q_values = range(0, 2)  # Seasonal moving average order
m_values = [12]         # Seasonal period

# Create all possible combinations of SARIMA parameters
param_combinations = list(itertools.product(p_values, 
                                            d_values, 
                                            q_values, 
                                            P_values, 
                                            D_values, 
                                            Q_values, 
                                            m_values))

# Initialize AIC with a large value
best_aic = float("inf")  
best_params = None

# Perform grid search
for params in param_combinations:
    order = params[:3]
    seasonal_order = params[3:]
    
    try:
        model = sm.tsa.SARIMAX(df_train_oil, 
                               order=order, 
                               seasonal_order=seasonal_order,
                               )
        result = model.fit(disp=False)
        aic = result.aic
        
        # Ensure the convergence of the model
        if not math.isinf(result.zvalues.mean()):
            print(order, seasonal_order, aic)
        
            if aic < best_aic:
                best_aic = aic
                best_params = params
                
        else:
            print(order, seasonal_order, 'not converged')

    except:
        continue

# Print the best parameters and AIC
print("Best Parameters:", best_params)
print("Best AIC:", best_aic)
```
