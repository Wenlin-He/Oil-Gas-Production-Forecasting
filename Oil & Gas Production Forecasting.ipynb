# Problem statement
**The main business problem presented involves enhancing the decision-making process for an Oil and Gas trading company that trades in production rights for Oil/Gas wells in the US. ## Problem 1
Enhancing Production Efficiency for Improved Return on Investment ROI
### Problem statement
What well characteristics are most influential in optimizing oil and gas production and efficiency to maximize ROI, while minimizing water production and associated costs?

### Understanding
To improve ROI, it's crucial to identify the well attributes that drive efficient production of oil and gas. The analysis will focus on two key areas:

**1.  Correlation Analysis**: Determine which well characteristics (e.g., depth, direction, type) have the strongest relationship with higher oil and gas output, as measured by LiquidsProd_BBL and GasProd_MCF.

**2.  Efficiency Metrics**: Develop a metric for "Hydrocarbon to Water Ratio" to assess the efficiency of hydrocarbon production relative to water byproduct. A favorable ratio suggests a more efficient extraction process with potential cost savings in water management.

## Problem 2
Analyzing Production Longevity and Decline
### Problem statement
How does the production of oil and gas change over time post-completion, and what implications does this have for predicting the economic lifespan of wells?
### Understanding
Analyzing how production metrics change as a function of time since completion can provide insights into the expected productive life of a well, helping in planning for investments, maintenance, or decommissioning.

## Problem 3
Analyzing environmental sustainability
### Problem statement
How can we enhance the environmental sustainability of well completion operations through improved water management practices?
### Understanding
Well completion processes, particularly those that involve significant water usage and chemical handling such as hydraulic fracturing and acidizing, pose environmental challenges. The comprehensive analysis of water usage, including volumes of water and chemicals like proppant and acid, can pinpoint opportunities to reduce the ecological footprint. Strategies could include optimizing resource consumption, innovating recycling and produced water-handling methods and ensuring adherence to environmental standards, all while maintaining operational efficiency.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

datapath = Path('data')
```

## Dataset 1


The meta data sheet primarily offers comprehensive details about the wells, including their identifiers, operators, well types, and geographic location. This information is crucial for categorizing wells and analyzing performance across different dimensions。

```python
cols_meta = ['API_UWI','StateWellType','WellPadDirection',
                            'ENVWellStatus','Trajectory','TVD_FT','MD_FT',
                            'Field','Proppant_LBS','TotalWaterPumped_GAL',
                            'TotalFluidPumped_BBL','AcidVolume_BBL',
                            'StateProvince','County','Latitude','Longitude']
metadata = pd.read_excel(datapath / 'Metadata.xlsx', usecols=cols_meta)
metadata = metadata.drop_duplicates()

metadata.head(5)
```

## Dataset 2

The production data sheet provides detailed monthly production figures, cumulative production metrics, and completion details. This data is vital for understanding production trends, forecasting future production, and assessing the economic viability of wells.

```python
# Import selected columns from the production data Excel file
cols_prod = ['API_UWI','TotalCompletionMonths','WellName','ProducingMonth','TotalProdMonths',
             'ProducingDays','Prod_BOE','Prod_MCFE','WaterProd_BBL'
            ,'LiquidsProd_BBL','GasProd_MCF','WaterProd_BBL'
             #,'CumProd_BOE','CumProd_MCFE','CumLiquids_BBL','CumGas_MCF','CumWater_BBL','ProducingOperator'
            ]
production = pd.read_excel(datapath / 'Production Data.xlsx', usecols=cols_prod)
production = production.drop_duplicates()
production.head(5)
```

## Dataset 3


```python
# Water data for enviromental related questions

df = pd.read_excel(datapath / 'c-115_produced_water_data_from_2014_operator_and_year.xlsx', skiprows=5)
df = df.dropna(axis=1, how='all')

# Find the first non-NaN row which should be the header
header_row_idx = df.first_valid_index()
# Set the correct header
df.columns = df.loc[header_row_idx]
# Remove the header row from the data
df = df.drop(header_row_idx)
# Reset the index of the DataFrame
df.reset_index(drop=True, inplace=True)
# Convert the column names to string type
df.columns = df.columns.astype(str)
# Remove the first row if it contains NaNs across all columns meant for data
df = df.dropna(subset=df.columns[:-1], how='all')
# Remove the last column if it is entirely NaN
if df.iloc[:, -1].isna().all():
    df = df.iloc[:, :-1]

water_data = df.copy()

water_data
```

## Problem 1

```python
# Merge production data with the meta data on API_UWI
merged_data = production.merge(metadata, on='API_UWI', how='left')
merged_data = merged_data.drop_duplicates().dropna()
merged_data.head(5)
```

```python
# key well characteristics that correlate with the oil and gas production
avg_liq_WellPadDirection = merged_data.groupby('WellPadDirection')['LiquidsProd_BBL'].mean().reset_index().sort_values('LiquidsProd_BBL', ascending=False)
avg_gas_WellPadDirection = merged_data.groupby('WellPadDirection')['GasProd_MCF'].mean().reset_index().sort_values('GasProd_MCF', ascending=False)

avg_liq_StateWellType = merged_data.groupby('StateWellType')['LiquidsProd_BBL'].mean().reset_index().sort_values('LiquidsProd_BBL', ascending=False)
avg_gas_StateWellType = merged_data.groupby('StateWellType')['GasProd_MCF'].mean().reset_index().sort_values('GasProd_MCF', ascending=False)

avg_TVD_MD = merged_data.groupby('API_UWI')[['TVD_FT','MD_FT','LiquidsProd_BBL','GasProd_MCF']].mean().reset_index()
```

```python
#Correlation Analysis: Determine which well characteristics (e.g., depth,
#direction, type) have the strongest relationship with higher oil and gas output
#, as measured by 'LiquidsProd_BBL' and 'GasProd_MCF'.

plt.figure(figsize=(10, 6))

# Bar plot for average LiquidsProd_BBL by StateWellType
plt.subplot(2, 2, 1)
sns.barplot(x='StateWellType', y='LiquidsProd_BBL', data=avg_liq_StateWellType)
plt.title('Average Liquid Production (BBL) by StateWellType')
plt.ylabel('Average Liquid Production (BBL)')
plt.xlabel('State Well Type')

# Bar plot for average GasProd_MCF by StateWellType
plt.subplot(2, 2, 2)
sns.barplot(x='StateWellType', y='GasProd_MCF', data=avg_gas_StateWellType)
plt.title('Average Gas Production (MCF) by StateWellType')
plt.ylabel('Average Production MCFE')
plt.xlabel('State Well Type')

# Bar plot for average LiquidsProd_BBL by WellPadDirection
plt.subplot(2, 2, 3)
sns.barplot(x='WellPadDirection', y='LiquidsProd_BBL', data=avg_liq_WellPadDirection)
plt.title('Average Liquid Production (BBL) by Well Pad Direction')
plt.ylabel('Average Liquid Production (BBL)')
plt.xlabel('Well Pad Direction')

# Bar plot for average GasProd_MCF by WellPadDirection
plt.subplot(2, 2, 4)
sns.barplot(x='WellPadDirection', y='GasProd_MCF', data=avg_gas_WellPadDirection)
plt.title('Average Gas Production (MCF) by Well Pad Direction')
plt.ylabel('Average Production MCFE')
plt.xlabel('Well Pad Direction')

plt.tight_layout()
plt.show()


# Scatter plots  for Depth (TVD_FT and MD_FT) vs Production Volumes
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
sns.scatterplot(x='TVD_FT', y='LiquidsProd_BBL', data=avg_TVD_MD)
plt.title('Liquid Production (BBL) by True Vertical Depth')

plt.subplot(2, 2, 2)
sns.scatterplot(x='TVD_FT', y='GasProd_MCF', data=avg_TVD_MD)
plt.title('Gas Production (MCF) by True Vertical Depth')


plt.subplot(2, 2, 3)
sns.scatterplot(x='MD_FT', y='LiquidsProd_BBL', data=avg_TVD_MD)
plt.title('Liquid Production (BBL) by Measured Depth')

plt.subplot(2, 2, 4)
sns.scatterplot(x='MD_FT', y='GasProd_MCF', data=avg_TVD_MD)
plt.title('Gas Production (MCF) by Measured Depth')

plt.tight_layout()
plt.show()
```

```python
# create variable 'Hydrocarbon_to_water_ratio'
merged_data = merged_data[merged_data['WaterProd_BBL'] != 0]
merged_data['Hydrocarbon_to_Water_Ratio'] = merged_data['LiquidsProd_BBL'] / merged_data['WaterProd_BBL']
merged_data = merged_data.dropna()
```

```python
merged_data['Hydrocarbon_to_Water_Ratio'].describe()
```

```python
# key well characteristics that correlate with the hydrocarbon/water ratio
# WellPadDirection
avg_HydroWater_WellPadDirection = merged_data.groupby('WellPadDirection')['Hydrocarbon_to_Water_Ratio'].mean().reset_index().sort_values('Hydrocarbon_to_Water_Ratio', ascending=False)
# StateWellType
avg_HydroWater_StateWellType = merged_data.groupby('StateWellType')['Hydrocarbon_to_Water_Ratio'].mean().reset_index().sort_values('Hydrocarbon_to_Water_Ratio', ascending=False)
# True vertical depth and measured depth
avg_HydroWater_TVD_MD = merged_data.groupby('API_UWI')[['TVD_FT','MD_FT','Hydrocarbon_to_Water_Ratio']].mean().reset_index()
```

```python
#Efficiency Metrics: Develop a metric for "Hydrocarbon to Water Ratio"
#to assess the efficiency of hydrocarbon production relative to water byproduct.
# A favorable ratio suggests a more efficient extraction process with potential cost savings in water management.

plt.figure(figsize=(10, 6))

# Bar plot for average Hydrocarbon_to_Water_Ratio by StateWellType
plt.subplot(1, 2, 1)
sns.barplot(x='StateWellType', y='Hydrocarbon_to_Water_Ratio', data=avg_HydroWater_StateWellType)
plt.title('Average Hydrocarbon_to_Water_Ratio by StateWellType')
plt.ylabel('Average Hydrocarbon_to_Water_Ratio')
plt.xlabel('State Well Type')


# Bar plot for average Hydrocarbon_to_Water_Ratio by WellPadDirection
plt.subplot(1, 2, 2)
sns.barplot(x='WellPadDirection', y='Hydrocarbon_to_Water_Ratio', data=avg_HydroWater_WellPadDirection)
plt.title('Average Hydrocarbon_to_Water_Ratio by Well Pad Direction')
plt.ylabel('Average Hydrocarbon_to_Water_Ratio')
plt.xlabel('Well Pad Direction')

plt.tight_layout()
plt.show()



# Continious variables: Vertical Depth, Measured Depth
plt.figure(figsize=(10, 6))

# Scatter plot for Hydrocarbon_to_Water_Ratio by True Vertical Depth
plt.subplot(1, 2, 1)
sns.scatterplot(x='TVD_FT', y='Hydrocarbon_to_Water_Ratio', data=avg_HydroWater_TVD_MD)
plt.title('Hydrocarbon_to_Water_Ratio by True Vertical Depth')
plt.xlabel('True Vertical Depth (feet)')
plt.ylabel('Hydrocarbon_to_Water_Ratio')

# Scatter plot for Hydrocarbon_to_Water_Ratio by Measured Depth
plt.subplot(1, 2, 2)
sns.scatterplot(x='MD_FT', y='Hydrocarbon_to_Water_Ratio', data=avg_HydroWater_TVD_MD)
plt.title('Hydrocarbon_to_Water_Ratio by Measured Depth')
plt.xlabel('Measured Depth (feet)')
plt.ylabel('Hydrocarbon_to_Water_Ratio')

plt.tight_layout()
plt.show()
```

```python
# Correlation matrix to see the pairwise relationships between all variables
data_matrix = merged_data[['TVD_FT', 'MD_FT','LiquidsProd_BBL', 'GasProd_MCF',
                          #'Proppant_LBS','TotalWaterPumped_GAL','TotalFluidPumped_BBL','AcidVolume_BBL'
                          ]]
correlation_matrix = data_matrix.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, vmin=-1, vmax=1, linewidths=0.05, cmap='viridis',square=True)
plt.title('Correlation Matrix of Well Characteristics and Production Volumes')
plt.show()
```
